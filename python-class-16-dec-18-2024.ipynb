# Newton's Method - Geometric

```python
def newton_v1(f,Df,x0,epsilon,max_iter):
    '''Approximate solution of f(x)=0 by Newton's method.

    Parameters
    ----------
    f : function
        Function for which we are searching for a solution f(x)=0.
    Df : function
        Derivative of f(x).
    x0 : number
        Initial guess for a solution f(x)=0.
    epsilon : number
        Stopping criteria is abs(f(x)) < epsilon.
    max_iter : integer
        Maximum number of iterations of Newton's method.

    Returns
    -------
    xn : number
        Implement Newton's method: compute the linear approximation
        of f(x) at xn and find x intercept by the formula
            x = xn - f(xn)/Df(xn)
        Continue until abs(f(xn)) < epsilon and return xn.
        If Df(xn) == 0, return None. If the number of iterations
        exceeds max_iter, then return None.

    Examples
    --------
    >>> f = lambda x: x**2 - x - 1
    >>> Df = lambda x: 2*x - 1
    >>> newton(f,Df,1,1e-8,10)
    Found solution after 5 iterations.
    1.618033988749989
    '''

    # records position of each point
    x_arr = []
    y_arr = []

    # starting point
    xi = x0

    
    for i in range(0,max_iter):
        # finds the corresponding y for xi
        fxi = f(xi)

        # records the point
        x_arr.append(xi)
        y_arr.append(fxi)

        # checking if it converges
        if abs(fxi) < epsilon:
            print('\nFound solution after',i+1,'iterations.')
            return xi, x_arr, y_arr
            
        Dfxi = Df(xi)

        # if it becomes flat at xi, it has no roots
        if Dfxi == 0:
            print('Zero derivative. No solution found.')
            return None
            
        xi = xi - fxi/Dfxi # this is the Newton's method eq.
        
    
    
    print('Exceeded maximum iterations. No solution found.')
    
    
    return None
```

```python
f = lambda x: x**2 - x - 1
Df = lambda x: 2*x - 1

#newton_v1(f,Df,x0,epsilon,max_iter)
newton_v1(f,Df,4,1e-8,10)[1] # why are we using [0], remember this
```

```python
import matplotlib.pyplot as plt
import numpy as np



results = newton_v1(f,Df,9,1e-9,10)
xi = results[0] # the solution point
x_arr = results[1] # x coordinates of the travel path
y_arr = results[2] # y coordinates of the travel path


x = np.linspace(0, 10, 500)


plt.figure(figsize=(12, 6))


plt.plot(x, f(x), lw=2, color='silver') # plots the solid line; lw is line width
plt.plot(xi, f(xi), 'r', marker='v', markersize=15) # plots the solution point
plt.scatter(x_arr, y_arr, s=25, alpha=0.8, color='b',label='Path to the Root')

#plt.autoscale(False) # To avoid that the scatter changes limits

plt.title("Finding roots using Newton's method", fontsize=18)
plt.xlabel(r"$x$", fontsize=18)
plt.ylabel(r"$f(x)$", fontsize=18)
#plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])
plt.ylim(0, 120)

plt.tight_layout()
plt.savefig('chapter3_Root_Finding_I.pdf')
```

# V2 function of Newton's method

```python
def newton_v2(f,Df,x0,epsilon,max_iter):

    xi = x0

    for i in range(0, max_iter):
        xi = xi - f(xi)/Df(xi) # Newton's method eq. applied to find the next point 
        fxi = f(xi) # finds the corresponding y
        Dfxi = Df(xi)
        
        if abs(xi - x0) < epsilon:  # this is the appropriate condition!!!
            # abs(xi - x0) is actually delta which we have used in grad descent
            print('\nFound solution after', i+1, 'iterations.')
            print('root = ', xi)
            return xi
        
        x0 = xi
        
        if Dfxi == 0:
            print('Zero derivative. No solution found.')
            print('Last value: xi = ', xi)
            return None
    
    if i == max_iter:    
        print('Exceeded maximum iterations. No solution found.')
        print('Last value: xi = ', xi)
        return None
```
