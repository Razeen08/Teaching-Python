```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 160) # we already know what it does

beta = 1 #just a parameter, we will see what it does

#func.1
def sigmoid(x,beta):
    return 1./(1 + np.exp(-beta*x)) #1./ means elementwise multiplication
#np.exp() means e (euler number)



#func.2 derivative of func.1
def sigmoidPrime(x,beta):
    return 1./(1+np.exp(-beta*x))*(1-1./(1+np.exp(-beta*x)))*beta



fig, axes = plt.subplots(3, 1, figsize=(10, 16))


# why are we using this?
def title_and_labels_and_legend(ax, title):
    ax.set_title(title)
    ax.set_xlabel('x', fontsize=16)
    ax.set_ylabel('f(x)', fontsize=16)
    ax.legend()
    ax.grid();


# plotting in sublot row 0
#plot(x, y)
y = sigmoid(x,beta)
p = axes[0].plot(x, y, '--r', label='sigmoid(x)')
title_and_labels_and_legend(axes[0], "Sigmoid Function - sig(x)") # performs multiples tasks defined in title_and_labels_and_legend(ax, title)


# plotting in sublot row 1
p = axes[1].plot(x, sigmoidPrime(x,beta), '--g', label='sigmoidPrime(x)')
title_and_labels_and_legend(axes[1], "SigmoidPrime Function - sigPrime(x)")


# plotting in sublot row 2
p = axes[2].plot(x, sigmoid(x,beta), '--r', label='sigmoid(x)')
p = axes[2].plot(x, sigmoidPrime(x,beta), '--g', label='sigmoidPrime(x)')
title_and_labels_and_legend(axes[2], "Sigmoid Function and its Prime")
plt.show()


#saving diff. types
fig.savefig('Sigmoid_and_its_PrimeHR.pdf', dpi=1000)
fig.savefig('Sigmoid_and_its_Prime.png', dpi=200)
fig.savefig('Sigmoid_and_its_Prime.jpg', dpi=200)

fig.savefig('Sigmoid_and_its_PrimetooLR.png', dpi=20)
fig.savefig('Sigmoid_and_its_PrimetooHR.png', dpi=1200)
```

```python
# importing necessary packages
import numpy as np
#import matplotlib as mpl
import matplotlib.pyplot as plt
# from mpl_toolkits import mplot3d; less specific, loads too many things, wastes memory
from mpl_toolkits.mplot3d.axes3d import Axes3D # more specific



fig, axes = plt.subplots(1, 2, figsize=(18, 8), subplot_kw={'projection': '3d'}) 



def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel("$x$", fontsize=18)
    ax.set_ylabel("$y$", fontsize=18)
    ax.set_zlabel("$z$", fontsize=18)
    ax.legend()
    ax.grid()



omega = 1

steps = 400
t = np.linspace(0, 40, steps)
radius = 2 - 0.075*t # radius decreases as t increases, the line goes near to the axis
p = axes[0].plot(radius*np.cos(omega*t), radius*np.sin(omega*t), 10-0.5*omega*t, '*--b', label='3D Plot')
title_and_labels(axes[0], "path of a drone - plot")





colors = ('b','g','r','c','y')  # tuple
c_list = [] #empty list

for c in colors:
    c_list.extend(c * int(steps/len(colors))) # appends 'steps/no_of_color' number of c into c_list



    
p = axes[1].scatter(radius*np.cos(omega*t), radius*np.sin(omega*t), 10-0.5*omega*t, c=c_list, label='Scatter')
title_and_labels(axes[1], "path of a drone - scatter")
axes[1].set_zlim((-2.5, 2.5)) # cutting the z axis to zoom


axes[0].text(1.2, 1.5, 7, "We start here ...")
axes[0].text(0.5, 0, 0, "... and end here")

fig.tight_layout()

#fig.savefig('3d_plot.pdf', dpi=200) # or plt.savefig('3d_plot.pdf', dpi=200)
#fig.savefig('3d_plot.png', dpi=200)
#fig.savefig('3d_plot.jpg', dpi=200)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# new import
from ipywidgets import interact

x = np.linspace(-10, 10, 160)

def sig(beta):
    y1 = 1./(1+np.exp(-beta*x))
    y2 = 1./(1+np.exp(-beta*x))*(1-1./(1+np.exp(-beta*x)))*beta
    plt.plot(x, y1, label='sig(x)')
    plt.plot(x, y2, label='sigPrime(x)')
    plt.ylim((-0.05, 1.05))
    plt.xlim((-10, 10))
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.legend()
    plt.show()

interact(sig, beta=(0, 2, 0.1));
```
