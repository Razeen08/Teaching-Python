```python
def sq_sum_func(x, y, z):
    sq_x = pow(x,2)
    sq_y = pow(y,2)
    sq_z = pow(z,2)
    sq_sum = sq_x + sq_y + sq_z
    return sq_sum

sq_sum1 = sq_sum_func(4, 5, 6)
print(sq_sum1)
```

```python
sq_sum_func = lambda x, y, z: pow(x,2) + pow(y,2) + pow(z,2) # define inline functions
# structure
# function_name = lambda input_arguments: math_formula

sq_sum1 = sq_sum_func(4, 5, 6)
print(sq_sum1)
```

```python
# i want both multiplication and sum from the function
def multiply_and_sum(num1 = 1, num2 = 1):
    mult = num1*num2
    return mult # writing return means the end of func.
    sum = num1 + num2
    return sum

x = multiply_and_sum(56, 80)

print(x)
```

```python
# i want both multiplication and sum from the function
def multiply_and_sum(num1 = 1, num2 = 1):
    mult = num1*num2
    yield mult # returns value but keep running the func
    sum = num1 + num2
    yield sum

x = multiply_and_sum(56, 80)

print(type(x))

for a in x:
    print(a)
```

```python
def fibonacci(a = 0, b = 1):
    while True:
        yield a
        a, b = b, a + b

f = fibonacci() # now this is called a generator; generator is basically a func that du=ynamically generates values

print(type(f))

for xyz in range(20):
    print(next(f))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.01)
y1 = np.exp(-x)
y2 = np.exp(0.1*x)

plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')
plt.xlabel('x values (0, 10)')
plt.ylabel('y values ')
plt.title('y = e^(-x)')
plt.legend(loc = 'center right', fontsize = 'large')
```
