# Multivariate Interpolation

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 50)
y = np.linspace(-10, 10, 50)

def f(x, y):
    return -np.exp(-(2.5*(x + 2.5)**2 + 0.25*(y - 5.5)**2)) + np.exp(-((x - 2.5)**2 + 1.25*(y + 1.5)**2))

# creating the 2d planar where x belongs to [-5,5], y belongs to [-10,10]
X, Y = np.meshgrid(x, y)
Z = f(X, Y) 


f_interpolated_linear = interpolate.interp2d(x, y, Z, kind='linear') #creates a linear interp func
f_interpolated_cubic = interpolate.interp2d(x, y, Z, kind='cubic') #creates a cubic interp func
f_interpolated_quintic = interpolate.interp2d(x, y, Z, kind='quintic') #creates a quintic interp func
```

**Testing the functions**

```python
print(f'Real func value: {f(-1.5, 1.5)}')
print(f'Linear interp func value: {f_interpolated_linear(-1.5, 1.5)}')
print(f'Cubic interp func value: {f_interpolated_cubic(-1.5, 1.5)}')
print(f'Quintic interp func value: {f_interpolated_quintic(-1.5, 1.5)}')
```

**Finer grids**



```python
x_fine = np.linspace(x.min(), x.max(), 2*x.size) # just doubles the grid, previously it was 50, now it is 100
y_fine = np.linspace(y.min(), y.max(), 2*y.size)

X_fine, Y_fine = np.meshgrid(x_fine, y_fine)
Z_fine = f(X_fine, Y_fine) 


f_interpolated_linear_fine = interpolate.interp2d(x_fine, y_fine, Z_fine, kind='linear') #creates a linear interp func
f_interpolated_cubic_fine = interpolate.interp2d(x_fine, y_fine, Z_fine, kind='cubic') #creates a cubic interp func
f_interpolated_quintic_fine = interpolate.interp2d(x_fine, y_fine, Z_fine, kind='quintic') #creates a quintic interp func
```

```python
print(f'Real func value: {f(-1.5, 1.5)}')
print(f'Linear interp func value: {f_interpolated_linear_fine(-1.5, 1.5)}')
print(f'Cubic interp func value: {f_interpolated_cubic_fine(-1.5, 1.5)}')
print(f'Quintic interp func value: {f_interpolated_quintic_fine(-1.5, 1.5)}')
```

**Plotting: let's make a 1 by 3 plot for linear, cubic and quintic respectively.**

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 6))


#linear interpolation
c1 = ax[0].pcolormesh(x_fine, y_fine, f_interpolated_linear_fine(x_fine, y_fine), cmap='plasma')
# plt.pcolormesh(x, y, Z, cmap='')
ax[0].set_title('Linear Interpolation', fontsize=12)
ax[0].set_xlabel(r'$x$', fontsize=12)
ax[0].set_ylabel(r'$y$', fontsize=12)
# show colorbar
cb = fig.colorbar(c1, ax=ax[0])
cb.set_label(r'$Z$', fontsize=12)

# rest 2: self study

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

**irregular (random spaced) grid**

```python
x_rand = np.random.randn(50)
y_rand = np.random.randn(50)

# random points are not sorted but for meshgrid we need x and y to be sorted
x_irr = np.sort(x_rand)
y_irr = np.sort(y_rand)

X_irr, Y_irr = np.meshgrid(x_irr, y_irr)


# define func
def f(x, y):
    return np.cos(np.pi*x/2)*np.exp(-0.5*y)

#function values in irregular grid
Z_irr = f(X_irr, Y_irr)
```

**regular grid, quintic interpolation**

```python
#creating finer frid, evenly spaced
x = np.linspace(-5, 5, 501)
y = np.linspace(-5, 5, 501)

X, Y = np.meshgrid(x, y)
Z = f(X, Y) # func values for this grid

# quintic interpolation
f_interp_quint = interpolate.interp2d(x, y, Z, kind='quintic')
```

**plot**

```python
# plotting function in irregular grid
fig, ax = plt.subplots(1, 2, figsize=(12,6))

irr = ax[0].pcolormesh(x_irr, y_irr, Z_irr, cmap='plasma')
ax[0].set_title('Function on irregular grid', fontsize=12)
ax[0].set_xlabel(r'$x$', fontsize=12)
ax[0].set_ylabel(r'$y$', fontsize=12)

cb = fig.colorbar(irr, ax=ax[0])
cb.set_label(r'$Z$', fontsize=10)

# plotting interpolation in reguklar grid
reg = ax[1].pcolormesh(x, y, f_interp_quint(x, y), cmap='plasma')
ax[1].set_title('Quintic interpolation on fine regular grid', fontsize=12)
ax[1].set_xlabel(r'$x$', fontsize=12)
ax[1].set_ylabel(r'$y$', fontsize=12)

cb = fig.colorbar(reg, ax=ax[1])
cb.set_label(r'$Z$', fontsize=10)

plt.show()
```

**3d plot**

```python
from mpl_toolkits.mplot3d import Axes3D

# create figure
fig = plt.figure(figsize=(12, 6))

# plotting function in irregular grid
ax0 = fig.add_subplot(121, projection='3d')
ax0.plot_surface(x_irr, y_irr, Z_irr, cmap='plasma')
ax0.set_title('Function on irregular grid', fontsize=12)
ax0.set_xlabel(r'$x$', fontsize=12)
ax0.set_ylabel(r'$y$', fontsize=12)
ax0.set_zlabel(r'$Z$', fontsize=12)

# plotting interpolation in regular grid
ax1 = fig.add_subplot(122, projection='3d')
ax1.plot_surface(x, y, f_interp_quint(x, y), cmap='plasma')
ax1.set_title('Quintic interpolation on fine regular grid', fontsize=12)
ax1.set_xlabel(r'$x$', fontsize=12)
ax1.set_ylabel(r'$y$', fontsize=12)
ax1.set_zlabel(r'$Z$', fontsize=12)

plt.show()
```
