# Set Methods and Numpy

```python
fruits = {"apple", "banana", "cherry"}

print(type(fruits))
```

```python
# two ways of defining a set
# 1 second brackets{}
names = {'Israq', 'Sakib', 'Razeen'}

# 2 using set() and instert a list
more_names = set(['Israq', 'Sakib', 'Razeen'])

print(type(names))
print(type(more_names))
```

```python
# set ignores duplicate

duplicate_names = {'Israq', 'Israq', 'Sakib', 'Razeen'}
print(duplicate_names)
```

```python
more_names = set(['Israq', 'Sakib', 'Razeen'])
more_names.add('Kamal')
more_names.add('Irfan')
print(more_names)

# no sort because is does not assign any indices
print(more_names[0]) # not allowed
```

```python
list5 = [30, 31, 32, 31, 35, 34, 33, 33, 34, 32, 31, 32, 32, 34, 36, 35] 
list5.sort(reverse=True)
print(list5)
```

```python
x = {'Israq', 'Israq', 'Sakib', 'Razeen'}
y = {'Razeen', 'Sakib', 'Irfan', 'Kamal', 'Israq'}

diff = x.difference(y) # shows elements of x that are not in y, alt. 'x not y'
rev_diff = y.difference(x)

print(diff)
```

```python
x = {'Israq', 'Israq', 'Sakib', 'Razeen'}
y = {'Razeen', 'Sakib', 'Irfan', 'Kamal', 'Israq'}

print(x.issubset(y)) #is x a subset of y?
print(y.issubset(x))

print(x.issuperset(y)) #is x a superset of y?
print(y.issuperset(x))
```
