# Euler Method



```python
import numpy as np
import matplotlib.pyplot as plt

# interval
a = 0
b = 8
# step size
h = 0.1
# number of points in the solution
n = int((b-a)/h)

# initial value
f0 = 0.5
x0 = a

# define Euler method function
def Euler(phi, xa, xb, finit, n):
    x = xa
    y = finit

    X = np.zeros([n+1]) # array of zeros to save the solution 
    f = np.zeros([n+1])

    X[0] = xa
    f[0] = finit

    for i in range(n):
        y = y + phi(x,y)*h
        x = x + h

        f[i+1] = y
        X[i+1] = x

    return [X, f]


# define the differential equation phi
def phi(x, y):
    dydx = -1.15*y + 4.5*np.exp(-0.25*x)
    return dydx

# numerical solution
X, f = Euler(phi, a, b, f0, n)

# exact solution
f_exact = (f0-5)*np.exp(-1.15*X) + 5*np.exp(-0.25*X)

# plot
plt.figure(figsize=(12, 6))
plt.scatter(X, f, label='Numerical Solution')
plt.plot(X, f_exact, 'r', label='Exact Solution')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
plt.title('Euler Method ODE Solution')
plt.grid()
plt.show()
```
