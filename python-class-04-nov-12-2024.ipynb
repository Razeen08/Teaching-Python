# Matrix Sum - same as mathematical sum


```python
list1 = [[1,2,3],[2,3,5],[1,1,3]]
list2 = [[-1,5,2],[9,-6,-5],[7,2,0]]

mat1 = np.matrix(list1) 
mat2 = np.matrix(list2) 
```

```python
list1+list2 # just merges the lists, no mathematical sum
```

```python
# sum
mat_sum = mat1 + mat2

print('Matrix 1:\n', mat1)
print('Matrix 2:\n', mat2)
print('Summation:\n', mat_sum)
```

```python
# sub
mat_sub = mat1 - mat2

print('Matrix 1:\n', mat1)
print('Matrix 2:\n', mat2)
print('Substraction:\n', mat_sub)
```

```python
arr1 = np.array(mat1)
arr2 = np.array(mat2)

elementwise = arr1*arr2 # * sign denotes elementwise multiplication only if they are arrays

print('Array 1:\n', arr1)
print('Array 2:\n', arr2)
print('Elementwise Multiplication:\n', elementwise)
```

```python
#mat_product = mat1*mat2 # * sign denotes matrix product

#other ways to perform matrix product
#mat_product = mat1@mat2 # here @ denotes matrix product
mat_product = mat1.dot(mat2) # uses dot() method

print('Matrix 1:\n', arr1)
print('Matrix 2:\n', arr2)
print('Matrix Product:\n', mat_product)
```

```python
#can also multiply scalar

5*mat1
```

```python
#pow5th = mat1*mat1*mat1*mat1*mat1
#pow5th = pow(mat1, 5) # pow() is not in numpy
pow5th = np.linalg.matrix_power(mat1, 5)

print('Mat 1, 5th Power:\n', pow5th)
```

# Transpose

```python
transposed_mat1 = mat1.transpose()

print('Matrix:\n', mat1)
print('Transposed:\n', transposed_mat1)
```

# Inverse

```python
inversed_mat1 = np.linalg.inv(mat1)

print('Inversed:\n', inversed_mat1)

# let's check
print('The product of mat1 and inversed_mat1 should be identity:\n', mat1@inversed_mat1)
```
