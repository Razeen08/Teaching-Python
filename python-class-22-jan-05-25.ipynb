# Development of Ordinary Differential Equations and its solution

```python
import numpy as np
from scipy.integrate import odeint # new here
import matplotlib.pyplot as plt


# constant and specific growth rate
alpha = 0.08


# define the growth function
def exponentialGrowth(X,t):
    N = X[0] # initial value or N_0; where X must be bacterial population wrt to time
    dNdt = alpha*N
    return [dNdt]


# defining initial value; number of bact at time = 0
Ninit = 10


# defining the simulation time
t = np.linspace(0,100,101)
sol = odeint(exponentialGrowth, Ninit, t) # odeint syntax (growth_func, initialValue, time_array)
N, = sol.T #tranposes for convenience in plotting


plt.figure(figsize=(12,6))
plt.plot(t, N, 'r', linewidth=1.5, label='Bacterial Population w.r.t. Time')
plt.xlabel('Time (hours)')
plt.ylabel('Number of Bacteria')
plt.legend()
plt.grid()
plt.savefig('bacterial_population.jpg', dpi=300)
plt.show()
```

# Interactive graph of Sigmoid func

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact # new here; for interactive graphs


#define an array
x = np.linspace(-10, 10, 501)


def sig(beta): # beta will be the variable in interactive graph
    y = 1./(1+np.exp(-beta*x))
    plt.plot(x, y, '--g', linewidth=1.5, label='Sigmoid(x)')
    plt.xlim(-12, 12)
    plt.ylim(-0.1, 1.1)
    plt.xlabel(r'$x$')
    plt.ylabel(r'$y$')
    plt.grid()
    plt.legend(loc='upper left')
    plt.show()


interact(sig, beta=(0, 2, 0.1))
```

# interactive graph of previous bacterial population problem

```python
import numpy as np
from scipy.integrate import odeint # new here
import matplotlib.pyplot as plt
from ipywidgets import interact


t = np.linspace(0,100,101)

def exponentialGrowth(X, t, alpha):
    N = X[0] # initial value or N_0; where X must be bacterial population wrt to time
    dNdt = alpha*N
    return [dNdt]

def expGrowthFunc(alpha):
    Ninit = 10
    sol = odeint(exponentialGrowth, Ninit, t, args=(alpha,)) # notice the use of args=(alpha,) here
    N, = sol.T

    plt.figure(figsize=(12,6))
    plt.plot(t, N, 'r', linewidth=1.5, label='Bacterial Population w.r.t. Time')
    plt.xlabel('Time (hours)')
    plt.ylabel('Number of Bacteria')
    plt.title('Exponential Growth')
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()

interact(expGrowthFunc, alpha=(0, 0.1, 0.01))
```
