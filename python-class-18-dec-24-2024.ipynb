# Linear Interpolation (interp1d)

Giver data points are (-1,1), (0,3), (1,-4.5) ,..., (6,5.5)

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt



# define arrays for x and y coordinates of different datapoints
x = np.array([-1, 0, 1, 2, 3, 4, 5, 6])
y = np.array([1, 3, 4.5, 4, 3, 3.25, 4, 5.5])



# perform interpolation (creates a function)
f = interpolate.interp1d(x, y, kind='linear')


# for plotting the predicted line
xplot = np.linspace(min(x),max(x),150) 
yplot = f(xplot)


fig = plt.figure(figsize=(8, 5))
plt.scatter(x, y, s=55, alpha=0.7, color='b', label='Data Points')
plt.plot(xplot, yplot, 'r--', lw=1.5, label='Linear Interpolation')

plt.legend(loc='best')
plt.grid()
plt.xlabel(r"$x$", fontsize=18)
plt.ylabel(r"$y$", fontsize=18)

fig.tight_layout()
fig.savefig('chapter3-Handling-Data-Interpolation-Linear-Interpolation.png');
```

# Polynomial and its roots

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
#from numpy import polynomial as plnl # can be written as np.polynomial alternatively


# new here
from scipy import linalg


# create a 3rd order func.
pol_func_1 = np.polynomial.Polynomial([1,-2,-3,1])
pol_func_1
```

```python
# finds the roots of the func.
pol_func_1.roots()
```

```python
# prints the coeffs of func.
pol_func_1.coef
```

**interpolation**

```python
# root points (xRoots, yRoots)
xRoots = pol_func_1.roots()
yRoots = np.array([0,0,0])


# polynomial func in plot
xPlot = np.linspace(-2,5,150)
yPlot = pol_func_1(xPlot)


# plot
fig = plt.figure(figsize=(8, 4))

plt.plot(xPlot, yPlot, 'b', lw=2, label='Polynomial')
plt.scatter(xRoots, yRoots, s=55, alpha=0.7, color='r', label='Roots')

plt.legend(loc='best')
plt.grid()
plt.ylim([-10,15])
plt.ylabel(r"$p(x)$", fontsize=18)
plt.xlabel(r"$x$", fontsize=18)

plt.show() # suppress other output
```
