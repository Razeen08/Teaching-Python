# Sum - Left/Right/Middle

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-0.25*x**2)

a = 0
b = 5
n = 15

k = 10 # Factor to make the function more smoothly

x = np.linspace(a,b,n+1)
y = f(x)

X = np.linspace(a,b,k*n+1)
Y = f(X)

h = (b-a)/n
x_left = np.linspace(a,b-h,n)
x_midpoint = np.linspace(a+h/2,b - h/2,n)
x_right = np.linspace(a+h,b,n)

# integrating numerically (creating left, midpoint and right sum)
l_sum = np.sum(f(x_left) * h)
mp_sum = np.sum(f(x_midpoint) * h)
r_sum = np.sum(f(x_right) * h)

# plotting
with plt.xkcd():
    plt.figure(figsize=(15,15))

    plt.subplot(3,1,1)
    plt.plot(X,Y,'gray')
    x_left = x[:-1] # Left endpoints of the intervals
    y_left = y[:-1]
    plt.plot(x_left,y_left,'r.',markersize=8)
    plt.bar(x_left,y_left,width=(b-a)/n,color = 'green', 
            alpha=0.1,align='edge',edgecolor='b',
            label='Sum = {}'.format(l_sum))
    plt.title('Riemanns Left Sum, n = {}'.format(n))
    plt.ylabel('f(x)')
    plt.legend()

    plt.subplot(3,1,2)
    plt.plot(X,Y,'gray')
    x_mid = (x[:-1] + x[1:])/2 # Midpoints of the intervals
    y_mid = f(x_mid)
    plt.plot(x_mid,y_mid,'r.',markersize=8)
    plt.bar(x_mid,y_mid,width=(b-a)/n,color = 'silver', 
            alpha=0.1,edgecolor='b',
            label='Sum = {}'.format(mp_sum))
    plt.title('Riemanns Midpoint Sum / Midpoint Method, n = {}'.format(n))
    plt.ylabel('f(x)')
    plt.legend()

    plt.subplot(3,1,3)
    plt.plot(X,Y,'gray')
    x_right = x[1:] # Right endpoints of the intervals
    y_right = y[1:]
    plt.plot(x_right,y_right,'r.',markersize=8)
    plt.bar(x_right,y_right,width=-(b-a)/n,color = 'orange', 
            alpha=0.1,align='edge',edgecolor='b',
            label='Sum = {}'.format(r_sum))
    plt.title('Riemanns Right Sum, n = {}'.format(n))
    plt.ylabel('f(x)')
    plt.legend()

    plt.show()

print("Summary: Riemanns sum with",n,"subintervals:\n")
print("Left Sum:\t",l_sum,"\nMidpoint Sum:\t",mp_sum,"\nRight Sum:\t",r_sum)
print("\nWe see:\nLeft Sum > Midpoint Sum > Right Sum")
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Define function and parameters
def f(x):
    return np.exp(-0.25 * x**2)



a, b = 0, 5  # Interval
n = 25       # Number of subintervals or bars


# red dots
x = np.linspace(a, b, n + 1)
y = f(x)


# solid line
p = np.linspace(a, b, 10 * n + 1)
q = f(p)


h = (b - a) / n #width of each bar

#__________________________________ upto this part it will be same for all three type of sums



x_left = np.linspace(a, b - h, n)  # Left endpoints

# Calculate Left Sum
left_sum = np.sum(f(x_left) * h) # length = f(x_left) and h=width



# Plot Left Sum
plt.figure(figsize=(10, 5))
plt.plot(p, q, 'gray') #solid line in gray
x_left = x[:-1]
y_left = y[:-1]
plt.plot(x_left, y_left, 'r.', markersize=8) # plots the left points in red dot
plt.bar(x_left, y_left, width=h, color='green', alpha=0.5, edgecolor='b', align='edge', label=f'Sum = {left_sum:.5f}')
plt.title(f"Riemann's Left Sum, n = {n}")
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

# Right sum

```python
# Calculate Right Sum
x_right = np.linspace(a + h, b, n)  # Right endpoints
r_sum = np.sum(f(x_right) * h)

# Plot Right Sum
plt.figure(figsize=(10, 5))
plt.plot(X, Y, 'gray')
x_right = x[1:] # leaves out the most left point
y_right = y[1:]
plt.plot(x_right, y_right, 'r.', markersize=8)
plt.bar(x_right, y_right, width=-h, color='orange', align='edge', alpha=0.4, edgecolor='k', label=f'Sum = {r_sum:.5f}')
plt.title(f"Riemann's Right Sum, n = {n}")
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

# Mid sum

**is a bit different from left/right**

```python
# Calculate Midpoint Sum
x_mid = (x[:-1] + x[1:]) / 2  # Midpoints is the average of all right points and left points
y_mid = f(x_mid)
mp_sum = np.sum(f(x_mid) * h)

# Plot Midpoint Sum
plt.figure(figsize=(10, 5))
plt.plot(X, Y, 'gray')
plt.plot(x_mid, y_mid, 'r.', markersize=8)
plt.bar(x_mid, y_mid, width=h, color='silver', alpha=0.3, edgecolor='k', label=f'Sum = {mp_sum:.5f}')
# we don't need align='edge' in this case
plt.title(f"Riemann's Midpoint Sum, n = {n}")
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
