# Data Load

```python
file_path = '/kaggle/input/manipulated-iris-dataset-for-beginners/iris_dirty.csv'
df = pd.read_csv(file_path, header=None)
```

# Add Headers

```python
header = ['SpLn', 'SpWd', 'PtLn', 'PtWd', 'Species']

df.columns = header
df.head()
```

# Add Case ID

```python
# Adding case_id as the first column
df.insert(0, "Case ID", range(1, len(df) + 1))
```

```python
df.head()
```

```python
int(df['PtWd'][0][:-3])/10
```

# Column PtWd issue

```python
WdList = []

for i in range(0,len(df)):
    width = int(df['PtWd'][i][:-3])/10
    WdList.append(width)
```

```python
df_copy = df

df_copy.head()
```

```python
df_copy['PtWd2'] = WdList

df_copy.tail(10)
```

```python
df_copy.drop('PtWd2', axis=1, inplace=True)

df_copy.head(10)
```

```python
df = df_copy

df
```

# Save Prepared DataFrame

```python
df.to_csv('iris_clean.csv', index=False)
```
