# handling complex number

```python
z1 = 6 + 8j #complex number, real part 6, imaginary part +8
z2 = 5 - 4j

print(z1 + z2)
print(z1.real + z2.real)
print(z1.imag + z2.imag)

print(z1*z2)

print(z1.conjugate())
```

# Conditions

```python
x = 55
y = 51

if x < y:
    print('x is smaller')
elif x == y:
    print('They are equal!')
else:
    print('x is larger')
```
