```python
def my_func(x, y):
    return (3/2)*(x+2)**2 - (5/2)*(y-2)**2

# use case
my_func(5, 7)
```

```python
# anonymous func syntax: 
# lambda input_args: return

my_func_anom = lambda x, y: (3/2)*(x+2)**2 - (5/2)*(y-2)**2

my_func_anom(5, 7)
```

# Root of square sum

```python
x = np.array([1, 2, 2])

sq = np.array([i*i for i in x])

sq_sum = sq.sum()

root = np.sqrt(sq_sum)

root
```

```python
x = np.array([1, 2, 2])

rt = np.sqrt((x*x).sum())

rt
```
