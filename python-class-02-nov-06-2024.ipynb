```python
dict1 = {'12 PM': 30, '1 PM': 31, '2 PM': 32, '3 PM': 31, '4 PM': 35}
# dict has KEYS and their asscoiated VALUES
# VALUES can be anything, int, string, list, tuple
print(type(dict1))
```

```python
#method fromkeys() is used for creating a dictionary

heading_keys = ('Name', 'Number')

info_dict = dict.fromkeys(heading_keys)

print(info_dict)
```

```python
# we accessed list values with index, we will access dict values with key
# list[0]
#dict['Name']

print(info_dict['Name'])
```

```python
info_dict['Name'] = ['Israq', 'Sakib', 'Razeen']
info_dict['Numbers'] = [52, 46, 34]

print(info_dict)
```

```python
#we made a mistake, lets drop 'Number'

info_dict.pop('Number')

print(info_dict)
```

```python
#update()

info_dict.update({'Location': ['Germany', 'Germany', 'Bangladesh']})

print(info_dict)
```

```python
#info_dict.update({'Location': 'Netherlands'})
#print(info_dict)
# so update is limited
```

```python
info_dict['Location'][1] = 'Netherlands' # i have updated Sakib's location

print(info_dict)
```

```python
info_list = [['Israq', 'Sakib', 'Razeen'],
            [52, 46, 34],
            ['Germany', 'Germany', 'Bangladesh']]

info_list[2][1] = 'Netherlands' #using index instead of key when it comes to list

print(info_list)
```
