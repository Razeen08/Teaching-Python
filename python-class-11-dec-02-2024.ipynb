# Constructor

```python
# using constructor, you can define attricutes of an object while creating

class student:

    # constructor
    def __init__ (self, first_name, last_name, birthdate, birthmonth, courses, first_course, second_course):
        self.first_name = first_name
        self.last_name = last_name
        self.birthdate = birthdate
        self.birthmonth = birthmonth
        self.courses = courses
        self.first_course = first_course
        self.second_course = second_course

    def name(self):
        student_name = self.first_name + ' ' + self.last_name
        return student_name

    def birthday(self):
        print(f"Student's birthday: {self.birthmonth} {self.birthdate}")
```

```python
israq = student('Israq', 'Noor', 4, 'May', 5, 'Scientific Programming', 'Discrete Math') # created student object israq

full_name = israq.name()
print(f"Student's full name: {full_name}")
israq.birthday()
```

```python
import math

class circle:
    def __init__ (self, radius):
        self.radius = radius
    
    def area(self):
        circle_area = math.pi*self.radius**2
        return circle_area
        
    def perimeter(self):
        circle_perimeter = 2*math.pi*self.radius
        return circle_perimeter
```

```python
radius = int(input("Input the radius of the circle:"))


c2 = circle(radius)

print(f'Area of the circle: {c2.area()}')
print(f'Perimeter of the circle: {c2.perimeter()}')
```
