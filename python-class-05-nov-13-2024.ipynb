# Solving a system of linear equations

```python
b = [12, 1]
b = np.array(b).reshape(2,1) #is an array
# A is also an array

#v = np.linalg.inv(A)*b # do not use * as both A and b are arrays, it performs elementwise mult.
v = np.linalg.inv(A)@b

print(v)
```

```python
# eq1.... x + 5y = 11
# eq2.... 2x - 3y = 5

A1 = [[1, 5], [2, -3]]
A1 = np.array(A1)

b1 = [11, 5]
b1 = np.array(b1).reshape(2,1)

sol = np.linalg.inv(A1)@b1

print(sol)
```

```python
# eq1.... x + 5y + z = 14
# eq2.... 2x - 3y - 3z = -3
# eq3.... -5x + 7y + 2z = 3

A2 = [[1, 5, 1], [2, -3, -3], [-5, 7, 2]]
A2 = np.array(A2)

b2 = [14, -3, 3]
b2 = np.array(b2).reshape(3,1)

sol2 = np.linalg.inv(A2)@b2
#sol2 = np.linalg.solve(A2,b2) #alternative way

print(sol2)
```
