```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Data
dict = {
    'Dhaka': 36433505,
    'Chittagong': 29145000,
    'Sylhet': 9807000,
    'Rajshahi': 12807000,
    'Khulna': 18485858,
    'Barisal': 17485858,
    'Rangpur': 8325666,
    'Mymensingh': 11370000
}

cities = list(dict.keys())
population = list(dict.values())

# Discretize the viridis colormap
colors = cm.viridis(np.linspace(0, 1, len(cities)))

# Plot
plt.figure(figsize=(8, 8))
plt.pie(population, labels=cities, colors=colors, autopct='%1.1f%%', startangle=0)
plt.title("Population Distribution by City", fontsize=16)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,50,101) # 101 values from 0 to 50 (incl.)

# parameters:
Uo = 5 
Rc = 10
C = 1

# functions
U = Uo*(1-np.exp(-t/Rc*C))

plt.figure(figsize=(10,5))
plt.plot(t, U, '-b', label = 'Capacitor Load')
plt.xlabel('t [sec]')
plt.ylabel('U [V]')
plt.grid(linestyle = '--', alpha = 0.7)
plt.legend(loc='best')
plt.show()
```

# Sigmoid and its derivative

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 160) # we already know what it does

beta = 1 #just a parameter, we will see what it does

#func.1
def sigmoid(x,beta):
    return 1./(1+np.exp(-beta*x))


#func.2 derivative of func.1
def sigmoidPrime(x,beta):
    return 1./(1+np.exp(-beta*x))*(1-1./(1+np.exp(-beta*x)))*beta



fig, axes = plt.subplots(3, 1, figsize=(10, 16))


# why are we using this?
def title_and_labels_and_legend(ax, title):
    ax.set_title(title)
    ax.set_xlabel('x', fontsize=16)
    ax.set_ylabel('f(x)', fontsize=16)
    ax.legend()
    ax.grid();


# plotting in sublot row 0
p = axes[0].plot(x, sigmoid(x,beta), '--r', label='sigmoid(x)')
title_and_labels_and_legend(axes[0], "Sigmoid Function - sig(x)")


# plotting in sublot row 1
p = axes[1].plot(x, sigmoidPrime(x,beta), '--g', label='sigmoidPrime(x)')
title_and_labels_and_legend(axes[1], "SigmoidPrime Function - sigPrime(x)")


# plotting in sublot row 2
p = axes[2].plot(x, sigmoid(x,beta), '--r', label='sigmoid(x)')
p = axes[2].plot(x, sigmoidPrime(x,beta), '--g', label='sigmoidPrime(x)')
title_and_labels_and_legend(axes[2], "Sigmoid Function and its Prime")
plt.show()


#saving diff. types
fig.savefig('Sigmoid_and_its_Prime.pdf', dpi=200)
fig.savefig('Sigmoid_and_its_Prime.png', dpi=200)
fig.savefig('Sigmoid_and_its_Prime.jpg', dpi=200)
```
