# Bitwise shift

```python
# bitwise shift

# dec 12 = bin 1100
```

```python
145 >> 1 

# dec 145 = bin 1 0 0 1 0 0 0 1
# >> 1 means shift 1 bit right
# after shifting => (front) 0 1 0 0 1 0 0 0 (back)
# right shift adds 0 in front, leaves bits from back
```

```python
12 << 2

# dec 12 = bin 1100
# << 2 means 2 bit left shift
# after shifting => (front) 110000 (back)
# left shift adds 0 to the back
```

```python
12 & 3 # performs bitwise and operation
```

# While loop

```python
#structure

#while (any condition):
    # any operation

# the operation continues as long as the condition is true
```

```python
number = 21 # first I took a random number

while number >=5 and number <=35:
    res = number % 5
    #print('Residual: ', res, ' for number = ', number)
    print(f'Residual {res} for number = {number}')
    number += 1
```
