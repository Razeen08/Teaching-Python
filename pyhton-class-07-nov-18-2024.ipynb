# For Loop

```python
linear_list = np.arange(10)
print(linear_list)

added_israq = []

for israq in linear_list: # assumes every element in linear_list is israq
    squared_israq = pow(israq,2)
    added_israq.append(squared_israq)
    print(israq)

print(added_israq)
```

```python
added_israq = [pow(israq,2) for israq in linear_list]
print(added_israq)
```

```python
linear_list = np.arange(1, 10)
print(linear_list)

primes = [1, 2, 3, 5, 7, 11, 13]
non_prime = [] #initiate a blank list

for int in linear_list[:7]:
    if int not in primes:
        non_prime.append(int)
    else:
        continue #proceed with next iteration

print(non_prime)
```

# Even or Odd

```python
decision = input('What do you want to see, type E for Even or O for Odd:')

num_list = np.arange(1, 101) # integers 1 to 100 


if decision == 'E': #if true, then execute the codes below
    for n in num_list:
        if n % 2 == 0: #true if n is even
            print(n)
        else:
            continue # nothing to do here, look at the next number
            
elif decision == 'O':
    for n in num_list:
        if n % 2 == 1: #true if n is odd
            print(n)
        else:
            continue

else:
    print('Your input is not defined!\nPlease enter either "E" or "O".')
```

```python
decision = input('What do you want to see, type E for Even (up to 60) or O for Odd (up to 51):')

num_list = np.arange(1, 101)


if decision == 'E':
    for n in num_list:
        if n % 2 == 0: #true if n is even
            print(n)
        elif n > 60:
            break # breaks out of the loop iteration, gives you better control to stop somewhere
        else:
            continue # nothing to do here, look at the next number
            
elif decision == 'O':
    for n in num_list:
        if n % 2 == 1: #true if n is odd
            print(n)
        elif n > 51:
            break
        else:
            continue

else:
    print('Your input is not defined!\nPlease enter either "E" or "O".')
```

# Why do we use functions?

```python
numbers1 = np.linspace(100, 500, 5)
numbers2 = np.arange(1,6)

print(numbers1)
print(numbers2)

def multiply_or_sum(num1, num2): #num1 and num2 are input arguments
    mult = num1*num2
    sum = num1 + num2
    if mult <= 1000:
        print(f'The multiplication of {num1} and {num2} is {mult}')
    else:
        print(f'The sum of {num1} and {num2} is {sum}')

    return 5

for a, b in zip(numbers1, numbers2): #iterates parallely in both numbers1 and numbers2
    multiply_or_sum(a, b)
```

```python
val = multiply_or_sum(a, b)
print(val)
```
